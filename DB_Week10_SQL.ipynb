# SQL commands to create **tables**

```python
# Setting up the Environment and Importing Necessary Libraries
import sqlite3
import pandas as pd
# Creating a Database
conn = sqlite3.connect('lab_session.db')
cursor = conn.cursor()
```

```python
# Command to create tables
commands = [
  """
  CREATE TABLE Products (
    ProductID INTEGER PRIMARY KEY,
    ProductName TEXT,
    Price REAL
  );
  """,
  """
  CREATE TABLE Customers (
    CustomerID INTEGER PRIMARY KEY,
    CustomerName TEXT,
    ContactNumber TEXT
  );
  """,
  """
  CREATE TABLE Sales (
    SaleID INTEGER PRIMARY KEY,
    ProductID INTEGER,
    CustomerID INTEGER,
    SaleDate DATE,
    SaleAmount REAL,
    FOREIGN KEY (ProductID) REFERENCES Products(ProductID),
    FOREIGN KEY (CustomerID) REFERENCES Customers(CustomerID)
  );
  """,
  """
  CREATE TABLE Inventories (
    InventoryID INTEGER PRIMARY KEY,
    ProductID INTEGER,
    StockQuantity INTEGER,
    FOREIGN KEY (ProductID) REFERENCES Products(ProductID)
  );
  """,
  """
  CREATE TABLE Suppliers (
    SupplierID INTEGER PRIMARY KEY,
    SupplierName TEXT,
    ContactNumber TEXT
  );
  """
]
# Execute the commands to create tables
for command in commands:
  cursor.execute(command)
```

# Windows Functions:

```python
'''Connect to the database (Assuming the database is already created
and populated)'''
conn = sqlite3.connect('lab_session.db')
# Window Functions:
# Objective: Calculate the running total and ranking of sales for each product.
# Running Total
running_total_query = """
SELECT
  ProductID,
  SaleDate,
  SaleAmount,
  SUM(SaleAmount) OVER (PARTITION BY ProductID ORDER BY SaleDate) AS
RunningTotal
FROM
  Sales;
"""
running_total_df = pd.read_sql_query(running_total_query, conn)
running_total_df
# Ranking
ranking_query = """
SELECT
  ProductID,
  SaleDate,
  SaleAmount,
  RANK() OVER (PARTITION BY ProductID ORDER BY SaleAmount DESC) AS
Rank
FROM
  Sales;
"""
ranking_df = pd.read_sql_query(ranking_query, conn)
ranking_df
#Common Table Expressions (CTEs)
#Objective: Create a CTE to calculate the total sales for each product
#and then query the CTE to find products with total sales greater than
#a specific amount.
# CTE
cte_query = """
WITH TotalSales_CTE AS (
  SELECT
    ProductID,
    SUM(SaleAmount) AS TotalSales
  FROM
    Sales
  GROUP BY
    ProductID
)
SELECT
  ProductID,
  TotalSales
FROM
  TotalSales_CTE
WHERE
  TotalSales > 40;
"""
# Close the database connection
conn.close()
conn = sqlite3.connect('lab_session.db')
cte_df = pd.read_sql_query(cte_query, conn)
cte_df
```

```python
# Task №1:
task1_query = """
WITH Task1_CTE AS (
  SELECT
    ProductID,
    SUM(SaleAmount) AS TotalSales,
    AVG(SaleAmount) AS AverageSales,
    MIN(SaleAmount) AS MinimumSales,
    MAX(SaleAmount) AS MaximumSales
  FROM
    Sales
  GROUP BY
    ProductID
)
SELECT
  ProductID,
  TotalSales,
  AverageSales,
  MinimumSales,
  MaximumSales
FROM
  Task1_CTE;
"""
conn = sqlite3.connect('lab_session.db')
task1_df = pd.read_sql_query(task1_query, conn)
task1_df
```

```python
conn.close()
```

```python
# Task №2:
task2_query = """
  SELECT
    ProductID,
    SUM(SaleAmount) AS TotalSales,
    RANK() OVER (ORDER BY SUM(SaleAmount) DESC) AS SalesRank
  FROM
    Sales
  GROUP BY
    ProductID
  ORDER BY
    SalesRank;
"""
conn = sqlite3.connect('lab_session.db')
task2_df = pd.read_sql_query(task2_query, conn)
task2_df
```
